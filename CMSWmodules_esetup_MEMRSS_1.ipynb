```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Enter raw data (500199.0_PREMIXUP15_PU50+PREMIXUP15_PU50 )
root = np.array([1236.31 , 1240.57 , 1189.89, 1224.77, 1268.63,1246.66])
cmssw = np.array([1405.35 , 1379.25, 1377.31, 1405.12, 1363.17, 1363.17, 1434.53 ])
```

```python
# Calculate the average
root_mean = np.mean(root)
cmssw_mean = np.mean(cmssw)
```

```python
# Calculate the standard deviation
root_std = np.std(root)
cmssw_std = np.std(cmssw)
```

```python
# Create lists for the plot
build = ['ROOT master', 'CMSSW PCMs']
x_pos = np.arange(len(build))
rts = [root_mean, cmssw_mean]
error = [root_std, cmssw_std]
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, rts, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('RSS measurements, MB')
ax.set_xticks(x_pos)
ax.set_xticklabels(build)
ax.set_title('RSS memory per one event for FastSim test, MB')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('fastSIM_MEMRSS.png')
plt.show()
```
